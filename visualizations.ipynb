```python
import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
resnet_34_two_class = pd.DataFrame([[1518, 162], [176, 1935]], index=['Animated', 'DUTS'], columns=['Animated', 'DUTS'])
resnet_fpn_two_class = pd.DataFrame([[1633, 47], [88, 2023]], index=['Animated', 'DUTS'], columns=['Animated', 'DUTS'])
resnet_fpn_adv_two_class = pd.DataFrame([[1565, 115], [44, 2067]], index=['Animated', 'DUTS'], columns=['Animated', 'DUTS'])
```

```python
with plt.style.context('default'):
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
    fig.set_figheight(5)
    fig.set_figwidth(18)
    sn.heatmap(resnet_34_two_class, ax=ax1, annot=True, fmt='d', square=True, cmap=cm.get_cmap('Greens_r'))
    ax1.set_title('Resnet34')
    sn.heatmap(resnet_fpn_two_class, ax=ax2, annot=True, fmt='d', square=True, cmap=cm.get_cmap('Greens_r'))
    ax2.set_title('Resnet34 + FPN')
    sn.heatmap(resnet_fpn_adv_two_class, ax=ax3, annot=True, fmt='d', square=True, cmap=cm.get_cmap('Greens_r'))
    ax3.set_title('Resnet34 + FPN with Advanced Decoder')
    fig.suptitle('Confusion Matricies for Two Class Resnet-Based Models', fontsize=18)
    plt.show()
```

```python
resnet_34_three_class = pd.DataFrame([[713,54,2],[178,574,69],[31,18,2062]], index=['Anime', 'Cartoon', 'DUTS'], columns=['Anime', 'Cartoon', 'DUTS'])
resnet_fpn_three_class = pd.DataFrame([[ 769,32,58],[ 243,550,28],[  25,7,2079]], index=['Anime', 'Cartoon', 'DUTS'], columns=['Anime', 'Cartoon', 'DUTS'])
resnet_fpn_adv_three_class = pd.DataFrame([[ 775,42,42],[ 165,622,34],[  49,10,2052]], index=['Anime', 'Cartoon', 'DUTS'], columns=['Anime', 'Cartoon', 'DUTS'])
```

```python
with plt.style.context('default'):
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
    fig.set_figheight(5)
    fig.set_figwidth(18)
    sn.heatmap(resnet_34_three_class, ax=ax1, annot=True, fmt='d', square=True, cmap=cm.get_cmap('Greens_r'))
    ax1.set_title('Resnet34')
    sn.heatmap(resnet_fpn_three_class, ax=ax2, annot=True, fmt='d', square=True, cmap=cm.get_cmap('Greens_r'))
    ax2.set_title('Resnet34 + FPN')
    sn.heatmap(resnet_fpn_adv_three_class, ax=ax3, annot=True, fmt='d', square=True, cmap=cm.get_cmap('Greens_r'))
    ax3.set_title('Resnet34 + FPN with Advanced Decoder')
    fig.suptitle('Confusion Matricies for Three Class Resnet-Based Models', fontsize=18)
    plt.show()
```
